```python
# Import Module
import os

# Folder Path
path = r"C:\Users\Sarthak M Desai\Yolov5_data_folder"

# Change the directory
os.chdir(path)

# Read text File

g= open("Yolov5_test_data_f.txt","w")

def read_text_file(file_path,file_name,g):
    with open(file_path, 'r') as f:
        print(file_name)
        #print(f.read())
        data_str = f.read()
        g.write(file_name + '\n' + data_str + '\n')
    f.close()


# iterate through all file

count = 1
for file in os.listdir():
    # Check whether file is in text format or not
    if "png.rf" in file:
        file_path = f"{path}\{file}"
        print(count)
        count+=1
        # call read text file function
        read_text_file(file_path,f"{file}",g)
```
