```python
textFile = open("Yolov5_test_data_f.txt")
outputFile = open("Yolov5_test_data_a.txt",'w')

out = ""
count = 1
for line in textFile:
    line = line.strip()
    if line == "":
        continue
    if "png.rf" in line:
        print(count)
        count+=1
        out = out.strip()
        outputFile.write(out+"\n")
        #print(out)
        out = ""
        out = out + "test/"+ line + ","
        set2.add(line)
    else:
        line = line.split()
        out = out + "["
        x = float(line[1]) 
        y = float(line[2])
        w = float(line[3])
        h = float(line[4])
        
        x1 = round((x-w/2)*416)
        y1 = round((y-h/2)*416)
        x2 = round((x+w/2)*416)
        y2 = round((y+h/2)*416)
        
        out = out + str(x1) + ", " + str(y1) + ", " + str(x2) + ", " + str(y2) + ", 1.0, 1.0, " + line[0] + "]"
outputFile.write(out)
outputFile.close()
```
